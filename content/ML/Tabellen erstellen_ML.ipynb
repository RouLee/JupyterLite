### Vorbereitung der Daten

Aber um mit Tabellen zu arbeiten, muss man zuvor noch ein Paket installieren:
Lass die nächsten beiden Zellen laufen.

```python
import pandas as  pd #Paket von Befehlen für die Datenverarbeitung
```

```python
datenLauf={"vorname" : ["Sari", "Raphael","Silvia","Charly"],
"strecke" : [15,42,60,42],
"alter" : [20,23,15,70],
"postleitzahl" : [6520,6045,8055,3994],
"geschlecht" : ["w","m","w","m"]}

df_Lauf=pd.DataFrame(datenLauf)
print(df_Lauf)
```

Der Befehl `pd.DataFrame(datenLauf)` macht aus einer Sammlung von Daten eine Tabelle, ein so genanntes **Data Frame**. Dazu benutzt es den Befehl aus der pd Library. Meist wird kurz df für dataFrame benutzt.

**Wichtig zu wissen:** Die Art und Weise wie wir die Daten eingegeben haben, `datenLauf={ "Spaltenname": [Liste], ...}` nennt man ein Dictionary. Dictionaries werden in Python auch sonst häufig verwendet.

**Exkurs** Man kann auch eine weitere Zeile hinzufügen, aber um die Bezeichnungen zu verstehen, müssen wir unsere Daten zuerst analysieren und einige Befehle lernen.


```python
df_Lauf.loc[len(df_Lauf)] = ['Jan', 42, 27, 8032, 'm']  #Hinzufügen
print(df_Lauf)
df_Lauf.drop(df_Lauf.index[len(df_Lauf)-1], inplace= True) #Wieder löschen
print(df_Lauf)
```

Um den Download etwas einfacher zu gestalten habe ich euch die beiden nötigen Schritte in einen neuen Befehl integriert. Der Downloadbefehl braucht zwei Argumente, die url und den gewünschten Namen. Gib Acht, dass du beide Argumente als String übergeben musst. Schreibe beim Namen am Ende, dass es ein csv file ist. Unten ist der Befehl schon programmiert und du musst ihn nur noch ausführen. 

**Hinweis:** Auf Computern (also nicht iPads) kannst du dein File auch via Drag and Drop in denselben Ordner ziehen in dem dein Notebook sich befindet und dann direkt den Befehl `df_name=pd.read("filename")` verwenden.

Der Downloadbefehl ist schon programmiert. Führe die nächsten zwei Zeilen einfach aus.

```python
import requests, os
def download_csv(url,filename): 
    #Testen ob Datei bereits vorhanden
    if os.path.isfile(filename):
        print(f"Datei {filename} existiert bereits.")
    else:
        response=requests.get(url)
        with open(filename,'wb') as fh:
            fh.write(response.content)

    #Die URL ist aus dem Internet, 
    #filename ist dein eigener Name, so heisst die Datei danach in deinem persönlichen Jupyterordner.
```

```python
download_csv("https://www.bista.zh.ch/basicapi/ogd/data_gymnasiale_maturitaetsquote_gebiete","maturquote.csv")
```

Mit Pandas (und im Hintergrund Matplotlib) können wir auch ganz einfach Grafiken erstellen um die Daten zu visualisieren. Ein Beispiel einer solchen Grafik ist ein Histogramm.

Es zählt wie viele Gemeinden eine Maturitätsquote in einem gewissen Bereich haben. Somit sieht man wie die Anzahl der Maturitätsquoten verteilt ist.

b) Wie funktioniert ein Histogramm?
Der Wertebereich wird in Bereiche unterteilt (z.B. 11-13%). Nun wird für jeden Bereich gezählt, wie viele Gemeinden eine Quote in diesem Bereich haben und entsprechend der Balken gezeichnet.
