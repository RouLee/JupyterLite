## Aufgabe 2.5: Grafik erstellen

Mit Pandas (und im Hintergrund Matplotlib) können wir auch ganz einfach Grafiken erstellen um die Daten zu visualisieren. 

Ein Beispiel einer solchen Grafik ist ein Histogramm.

Der folgende Code erstellt ein Histogramm der Maturitätsquoten.
Lass die folgende Code-Zelle laufen und beantworte dann folgende Fragen:


a) Wie interpretierst du das Histogramm?
Es zählt wie viele Gemeinden eine Maturitätsquote in einem gewissen Bereich haben. Somit sieht man wie die Anzahl der Maturitätsquoten verteilt ist.

b) Wie funktioniert ein Histogramm?
Der Wertebereich wird in Bereiche unterteilt (z.B. 11-13%). Nun wird für jeden Bereich gezählt, wie viele Gemeinden eine Quote in diesem Bereich haben und entsprechend der Balken gezeichnet.

c) Sticht eine Gemeinde mit einer Maturitätsquote von 49% aus den Daten heraus?
Ja, da die Meisten gemeinden eher um den Bereich 10-30% liegen, ist eine Gemeinde mit 49% eher speziell.

d) Wie würdest du die Verteilung beschreiben? Was beobachtest du? (Wo liegt der grösste Teil der Gemeinden, was sagt es uns über die Anzahl von Gemeinden mit sehr hohen Quoten und tiefen Quoten aus?)

Die meisten sind im Bereich 10-30%. Es gibt nochmals über 20 Gemeinden mit einer sehr tiefen Quote.
