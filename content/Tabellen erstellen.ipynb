## Aufgabe 0: Wie würdest du die Daten in einer Tabelle darstellen?

Notiere deine Antwort hier:

```python
import pandas as  pd #Paket von Befehlen für die Datenverarbeitung
```

```python
datenLauf={"vorname" : ["Sari", "Raphael","Silvia","Charly"],
"strecke" : [15,42,60,42],
"alter" : [20,23,15,70],
"postleitzahl" : [6520,6045,8055,3994],
"geschlecht" : ["w","m","w","m"]}

dLauf=pd.DataFrame(datenLauf)
print(dLauf)
```

Der Befehl `pd.DataFrame(datenLauf)` macht aus einer Sammlung von Daten eine Tabelle, eine so genanntes **Data Frame**. Dazu benutzt es den Befehl aus der pd Library. Meist wird kurz df für dataFrame benutzt.

**Wichtig zu wissen:** Die Art und Weise wie wir die Daten eingegeben haben, `datenLauf={ "Spaltenname": [Liste], ...}` nennt man ein Dictionary. Dictionaries werden in Python auch sonst häufig verwendet.

**Exkurs** Man kann auch eine weitere Zeile hinzufügen, aber um die Bezeichnungen zu verstehen, müssen wir unsere Daten zuerst analysieren und einige Befehle lernen.


```python
dLauf.loc[len(dLauf)] = ['Jan', 42, 27, 8032, 'm']  #Hinzufügen
print(dLauf)
dLauf.drop(dLauf.index[len(dLauf)-1], inplace= True) #Wieder löschen
print(dLauf)
```

Um den Download etwas einfacher zu gestalten habe ich euch die beiden nötigen Schritte in einen neuen Befehl integriert. Der Downloadbefehl braucht zwei Argumente, die url und den gewünschten Namen. Gib Acht, dass du beide Argumente als String übergeben musst. Schreibe beim Namen am Ende, dass es ein csv file ist.

**Hinweis:** Auf Computern (also nicht iPads) kannst du dein File auch via Drag and Drop in denselben Ordner ziehen in dem dein Notebook sich befindet und dann direkt den Befehl `df_name=pd.read("filename")` verwenden.

```python
import requests
def download_csv(url,filename): 
    response=requests.get(url)
    with open(filename,'wb') as fh:
        fh.write(response.content)

    #Die URL ist aus dem Internet, 
    #filename ist dein eigener Name, so heisst die Datei danach in deinem persönlichen Jupyterordner.
```

```python
download_csv("https://www.web.statistik.zh.ch/ogd/data/bista/ZH_Maturitaetsquote_Gemeinden_und_Kanton.csv","maturquote.csv")
```
